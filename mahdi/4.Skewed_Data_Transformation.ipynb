# 4. Skewed Data Transformation

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
%matplotlib inline
```

```python
train_file = 'house-prices-advanced-regression-techniques/train_outlier_removed.csv'
train_df = pd.read_csv(train_file)
for continuesCol in train_df:
    print(continuesCol)
    tempCol = train_df[continuesCol].dropna()
    colMax = tempCol.max()
    colMin = tempCol.min()
    colDelta = colMax - colMin
    print('Max: ' + str(colMax))
    print('Min: ' + str(colMin))
    print('Delta: ' + str(colDelta))
    
    
    plt.hist(tempCol, bins=20)
    
    plt.show()
    print('----------------------')
```
