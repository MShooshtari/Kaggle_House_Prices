```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
%matplotlib inline
import xgboost as xgb
from xgboost import XGBRegressor
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
from pathlib import Path
path = Path('/content/drive/My Drive')
train_file = path/'house-prices-advanced-regression-techniques/train_normalized.csv'
train_df = pd.read_csv(train_file)

test_file = path/'house-prices-advanced-regression-techniques/test_normalized.csv'
test_df = pd.read_csv(test_file)
```

```python
from sklearn.metrics import mean_squared_error

X = train_df.drop(['SalePrice'], axis=1)
Y = train_df['SalePrice']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Initializing the ANN
classifier = Sequential()
# Adding the input layer and the first hidden layer 
# Number of nodes in each layer: Average of number of input nodes and ouput nodes
# In our case, we have 42 input nodes and 2 classes, therfore 1 output node, so the average is 22 or 21.
numberOfInputNodes = len(list(X_train))
numberOfOutputNodes = 1
numberOfMiddleNodes = int((numberOfInputNodes + numberOfOutputNodes)/2)
classifier.add(Dense(input_dim=numberOfInputNodes, output_dim=numberOfMiddleNodes, activation='relu'))
classifier.add(Dense(output_dim=numberOfMiddleNodes, activation='relu'))
classifier.add(Dense(output_dim=numberOfMiddleNodes, activation='relu'))
classifier.add(Dense(output_dim=numberOfOutputNodes, activation='linear'))

# Compiling the ANN
classifier.compile(optimizer='adam', loss='mean_squared_error')

# Fitting the ANN to the training set
classifier.fit(X_train, Y_train, batch_size=100, nb_epoch=2000)

Y_pred_val = classifier.predict(X_test)
# Calculation of Mean Squared Error (MSE) 
mse = mean_squared_error(Y_test,Y_pred_val)
print(mse)
```

```python
Y_pred = classifier.predict(test_df)
```

```python
original_test_file = path/'house-prices-advanced-regression-techniques/test.csv'
original_test_df = pd.read_csv(original_test_file)
id_col = original_test_df['Id']
result_df = pd.concat([id_col, pd.DataFrame(Y_pred, columns=['SalePrice'])], axis=1)
result_df.to_csv('ANN_reg_WO_PCA.csv', index=False)
```
